# Collaborative Filtering Approach

```python
x = ratings_with_book_name.groupby("User-ID").count()
```

```python
x = ratings_with_book_name.groupby("User-ID")["Book-Rating"].count() > 200
```

```python
knowledgeble_users_df = ratings_with_book_name[ratings_with_book_name["User-ID"].isin(x)]
```

```python
knowledgeble_users_df
```

## we have the dataframe where 811 users are unique and have rated books over 200 meaning rated more than 200 books 

## Books which has atleast 50 rating or more

```python
y = knowledgeble_users_df.groupby("Book-Title")["Book-Rating"].count() >=50
```

```python
y = y[y].index
```

```python
famous_books_df = knowledgeble_users_df[knowledgeble_users_df["Book-Title"].isin(y)]
```

# user - book pivot table

```python
pivot = famous_books_df.pivot_table(index = "Book-Title", columns = "User-ID", values = "Book-Rating")
```

```python
pivot
```

```python
pivot.fillna(0, inplace = True)
```

```python
pivot
```

## Cosine Similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot)
```

```python
similarity_scores
```

```python
np.where(pivot.index == "Zoya")[0][0]
```

# Recommendation Function

```python
def recommend(book_name):
    similar_books = []  # empty list of books

    # finding the index of the given book in the pivot table
    index = np.where(pivot.index == book_name)[0][0]  

     # getting the index of top 5 similar books
    similar_books_indexes = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1], reverse = True)[1:6] 

    # going through the index to find the book name and add to similar_books list
    for i in similar_books_indexes:
        similar_books.append(pivot.index[i[0]])

    # finally return suggestions -> most similar books
    return similar_books
```

```python
print(recommend("Zoya"))
```

```python
 print(recommend("1984"))
```

```python
print(recommend("4 Blondes"))
```
